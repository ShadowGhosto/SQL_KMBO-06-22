## Задание 2
Для таблиц альбомов и плейлистов добавить столбец quantity - количество треков в них.

```python
%%sql
ALTER TABLE albums
ADD quantity INTEGER;

ALTER TABLE playlists
ADD quantity INTEGER;
```

```python
%%sql
-- Подсчет количества треков в каждом альбоме
UPDATE Albums
SET quantity = (SELECT COUNT(*) FROM Tracks WHERE Tracks.AlbumId = Albums.AlbumId);

-- Подсчет количества уникальных треков в каждом плейлисте
UPDATE Playlists
SET quantity = (SELECT COUNT(DISTINCT TrackId) FROM Playlist_Track WHERE Playlist_Track.PlaylistId = Playlists.PlaylistId);
```

## Задание 3
Объединить 3 таблицы - медиатипы, жанры и треки в одну общую таблицу tracks_upd. Заменить в БД три старые таблицы на эту новую. Вывести результат.

```python
%%sql
CREATE TABLE tracks_upd AS
SELECT 
    tracks.TrackId, 
    tracks.Name, 
    tracks.AlbumId, 
    media_types.Name AS MediaType, 
    genres.Name AS Genre, 
    tracks.Composer, 
    tracks.Milliseconds, 
    tracks.Bytes, 
    tracks.UnitPrice
FROM 
    tracks 
JOIN 
    media_types ON tracks.MediaTypeId = media_types.MediaTypeId
JOIN 
    genres ON tracks.GenreId = genres.GenreId;
```

```python
%%sql
DROP TABLE media_types;
DROP TABLE genres;
DROP TABLE tracks;
```

```python
%%sql
SELECT * FROM tracks_upd
```

## Задание 4
Провести промо-акцию: в получившейся таблице сделать скидку 20% на треки жанра Rock и 10% на треки без композитора. Если выполняются оба условия, то скидка всё ещё 20%. Добавить столбец SALE (true/false - есть акция или нет). Вывести 100 первых треков.

```python
%%sql
-- Добавление столбца SALE в таблицу tracks_upd
ALTER TABLE tracks_upd
ADD SALE BOOLEAN DEFAULT FALSE;
```

```python
%%sql
-- Обновление столбца SALE и UnitPrice для треков жанра Rock
UPDATE tracks_upd
SET SALE = TRUE, UnitPrice = UnitPrice * 0.8
WHERE Genre = 'Rock';

-- Обновление столбца SALE и UnitPrice для треков без композитора, которые не относятся к жанру Rock
UPDATE tracks_upd
SET SALE = TRUE, UnitPrice = UnitPrice * 0.9
WHERE Composer IS NULL AND Genre != 'Rock';

-- Вывод первых 100 треков
SELECT * FROM tracks_upd
```

```python
%%sql
-- Создать таблицу Cities
CREATE TABLE Cities (
    CityId INTEGER PRIMARY KEY,
    City TEXT NOT NULL,
    State TEXT,
    Country TEXT NOT NULL,
    FOREIGN KEY (CityId) REFERENCES Customers (CityId)
);
```

```python
%%sql
-- Заполнить таблицу Cities данными из таблицы Customers
INSERT INTO Cities (City, State, Country)
SELECT City, State, Country FROM CUSTOMERS;
```
